```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

df = pd.read_csv('dataset.csv')

print(df.head())
print(df.info())
print(df.isnull().sum())

df_reduced = df.sample(frac=0.5, random_state=1)

print(df_reduced.head())
print(df_reduced.info())
print(df_reduced.isnull().sum())

df_reduced['track_name'] = df_reduced['track_name'].astype(str)
analyzer = SentimentIntensityAnalyzer()

def sentiment_analysis(text):
    scores = analyzer.polarity_scores(text)
    return scores['compound']

df_reduced['sentiment_score'] = df_reduced['track_name'].apply(sentiment_analysis)
df_reduced['sentiment'] = df_reduced['sentiment_score'].apply(lambda x: 'positive' if x >= 0.05 else ('negative' if x <= -0.05 else 'neutral'))

df_reduced.to_csv('reduced_dataset.csv', index=False)

print(df_reduced['sentiment'].value_counts())
```
